# [Sourmashconsumr analysis ]( https://arcadia-research.pubpub.org/pub/resource-sourmashconsumr)


This section uses functions from sourmashconsumr to visualize outputs from sourmash to compare samples and to look at taxonomy & abundance over time in Illumina metagenomic data.

The 'sourmash sketch dna' command with -p flags k=31,scaled=1000 was used to make signatures for all metagenomic assemblies. The 'sourmash compare' command was then used to make a similarity matrix for the assemblies. The 'sourmash sketch dna' command with -p flags k=31,scaled=1000,abund was used to make signatures for paired-end Illumina read files. The 'sourmash gather' command with -k 31 and --scaled 1000 options was used for each of the nine Illumina read file signatures. For 'sourmash gather', the preprepared sourmash Genbank genomes from March 2022 k31 databases for viruses, bacteria, archaea, protozoa, and fungi, plus the custom 'cheesegenomes-k31-scaled1k' databases were used. The 'sourmash tax annotate' command was then used on the resulting gather files with the taxonomy sheets for these six databases. This resulted in nine sr.with-lineages.csv files.
[Sourmash version 4.6.1](https://joss.theoj.org/papers/10.21105/joss.00027) was used.

```r
#load libraries needed, source sourmashconsumr funcitons, and set a color scale
library(readr)
library(dplyr)
library(tidyr)
library(purrr)
library(tibble)
library(ggplot2)
library(ggthemes)
library(ggpubr)
library(ComplexUpset)
library(sourmashconsumr)

accent_v2 = c('#D1EADF','#5088C5', '#F28360', '#3B9886', '#F7B846', 
                       '#7A77AB', '#F898AE', '#C6E7F4', "#97CD78",'#F8C5C1',
                       '#B5BEA4', '#DCBFFC', "#8A99AD","#DAD3C7","#C85152","#BABEE0")
```

### Visualize sourmash compare output 

```r
#read in output from sourmash compare, plot with sourmashconsumr function

compare <- read_csv("../processed_data/csv_tables_for_plotting/newmdscompare0323.csv", show_col_types = F) %>%
  mutate(sample = colnames(.)) %>% # make column names into a column
  column_to_rownames("sample") %>% # set rownames to column names
  as.matrix() 
compare
compare_mds <- compare %>% 
  dist() %>%
  cmdscale() %>%
  as.data.frame() %>%
  rename(MDS1 = V1, MDS2 = V2) %>%
  rownames_to_column("sample")
compare_mds
```

```r
# Add metadata columns
compare_mds$tech=c("s","zl","zl","zl","zl","zl","zl","s","s","s","zl","zl","zl","s","s","s","s","s") #short or long read
compare_mds$age=c("early","zlate","middle","early","middle","early","zlate","middle","zlate","middle","zlate","middle","early","zlate","early","middle","early","zlate")
compare_mds$cheese=c("o","e","e","e","o","o","o","e","o","w","w","w","w","w","w","o","e","e")
compare_mds
```

```r
# make MDS plot
options(repr.plot.width=10, repr.plot.height=10)

ggplot(compare_mds, aes(x = MDS1, y = MDS2, color=cheese, shape=age, size=tech)) +
  geom_point() + 
  theme_classic()+
scale_color_manual(values = c("#5088C5","#F28360","#97CD78"))+scale_shape_manual(values=c(15,17,18))+scale_size_manual(values=c(2,4))

```

### Look at taxonomy over time

```r
#Import timedf file listing query_names and their relative ages
timedf=suppressWarnings(read.csv("../processed_data/csv_tables_for_plotting/whitime.csv"))
timedf
#Use Sys.glob to make taxdf file from the three tax annotate files for cheese WH
taxdf=read_taxonomy_annotate(Sys.glob("../processed_data/csv_tables_for_plotting/whi*sr.with-lineages.csv"),intersect_bp_threshold=50000)
#Use Sys.glob to make taxdf file from the three tax annotate files for cheese WH
options(repr.plot.width=15, repr.plot.height=7)
plot=plot_taxonomy_annotate_ts_alluvial(taxdf,timedf,tax_glom_level="species",palette = accent_v2,fraction_threshold = 0.01,
                                   label = F)
plot+theme(legend.position ="bottom")
```
